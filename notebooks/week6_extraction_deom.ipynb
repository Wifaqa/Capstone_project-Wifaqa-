# Resume Parsing Pipeline Test

This notebook demonstrates the end-to-end extraction pipeline for the AI Recruiter project.

Pipeline:
PDF → Extracted Text → Structured JSON → Normalized Skills



```python
import chromadb
from chromadb.utils import embedding_functions
from pathlib import Path

session_dir = Path("data/sessions/Data Science")
index_dir = session_dir / "index"

embedding_function = embedding_functions.OpenAIEmbeddingFunction(
    model_name="text-embedding-3-small"
)

client = chromadb.Client(
    chromadb.config.Settings(
        persist_directory=str(index_dir)
    )
)

collection = client.get_or_create_collection(
    name="resumes",
    embedding_function=embedding_function
)

print("Document count:", collection.count())
```
